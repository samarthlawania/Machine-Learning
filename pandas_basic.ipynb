```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [7,4,3,5,6,2]
plt.plot(x,y,color="green",linestyle="--",marker="o",linewidth='2')
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
categories = ["a","b","c","d"]
values = [33,22,34,32]
plt.bar(categories,values,color="skyblue")
plt.title("Bar plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```
